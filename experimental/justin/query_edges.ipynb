## queryTargetInstances

```typescript
const mediaId = "80b5336730d3421790fffb490121fc3a"; // (1) ONE PREVIEW
// const mediaId = "c4c8e0dbca0946508f16afbf4dd168ad"; // (0) NO PREVIEW
// const mediaId = "b1821998c0e54e5695df7805969d6f59"; // (2) TWO PREVIEWS
```

```typescript
import { getJupyterCurrentViewer } from "infra/lib/jupyterUtils.ts";
const cv = await getJupyterCurrentViewer();
```

```typescript
import { BfEdge } from "packages/bfDb/coreModels/BfEdge.ts";
import { BfMedia } from "packages/bfDb/models/BfMedia.ts";
import { BfMediaNodeVideo } from "packages/bfDb/models/BfMediaNodeVideo.ts";
import { BfMediaNodeVideoGoogleDriveResource } from "packages/bfDb/models/BfMediaNodeVideoGoogleDriveResource.ts";
import type {
  BfGid,
  BfSid,
  BfTid,
} from "packages/bfDb/classes/BfBaseModelIdTypes.ts";
const media = await BfMedia.find(cv, mediaId);
const targets = await media.queryTargetInstances(
    BfMediaNodeVideoGoogleDriveResource,
);
```

```typescript
// rebuild queryTargetInstances
const TargetClassA = BfMediaNodeVideoGoogleDriveResource as unknown as typeof BfNode;
const sourceBfGidA = mediaId;
// const edgePropsToQuery = {};
const propsToQueryA = { };
console.log("queryTargets", TargetClassA, sourceBfGidA);

const targetEdgesA = await BfEdge.query(
  cv,
  { bfSid: sourceBfGidA, bfTClassName: TargetClassA.name },
  propsToQueryA,
);
console.log("targetEdges", targetEdgesA);
const targetEdgeIdsA = targetEdgesA.map((edge) => edge.metadata.bfTid)
  .filter(Boolean) as Array<BfTid>;
console.log("targetEdgeIds", targetEdgeIdsA);
const targetsA = await TargetClassA.query(
  cv,
  {},
  propsToQueryA,
  targetEdgeIdsA,
);
console.log(`expected 1 got ${targetsA.length}`)
```

```typescript
const bfMediaNodeVideoGoogleDriveResource = targets[0];
const bfMediaNodeVideoGoogleDriveResourceId = bfMediaNodeVideoGoogleDriveResource.metadata.bfGid;
// bfMediaNodeVideoGoogleResource
// targets.length
```

```typescript
// test if the source on the resource is the media
const sourceMedia = await bfMediaNodeVideoGoogleDriveResource.querySourceInstances(
    BfMedia,
)
const sourceMediaId = sourceMedia[0].metadata.bfGid;
console.log(sourceMediaId === mediaId, sourceMediaId, mediaId)
```

```typescript
const targetVideos = await bfMediaNodeVideoGoogleDriveResource.queryTargetInstances(
    BfMediaNodeVideo,
)
console.log(`expected 0 got ${targetVideos.length}`)
```

```typescript
// rebuild queryTargetInstances
const TargetClass = BfMediaNodeVideo as unknown as typeof BfNode;
const sourceBfGid = bfMediaNodeVideoGoogleDriveResourceId;
// const edgePropsToQuery = {};
const propsToQuery = { role: "PREVIEW" };
console.log("queryTargets", TargetClass, sourceBfGid);

const targetEdges = await BfEdge.query(
  cv,
  { bfSid: sourceBfGid, bfTClassName: TargetClass.name },
  propsToQuery,
);
console.log("targetEdges", targetEdges);
const targetEdgeIds = targetEdges.map((edge) => edge.metadata.bfTid)
  .filter(Boolean) as Array<BfTid>;
console.log("targetEdgeIds", targetEdgeIds);
const targets = await TargetClass.query(
  cv,
  {},
  propsToQuery,
  targetEdgeIds,
);
console.log(`expected 0 got ${targets.length}`)
```

```typescript
// Test if the source is the resource to on the targets returned
console.log("Source id should be bfMediaNodeVideoGoogleDriveResourceId")
for (const target of targets) {
    const source = await target.querySourceInstances(
        BfMediaNodeVideoGoogleDriveResource,
    )
    if (source.length === 0) {
        console.log("No source")
    } else {
        const sourceId = source[0].metadata.bfGid;
        const equality = sourceId === bfMediaNodeVideoGoogleDriveResourceId ? "✅" : "❌"
        console.log(`${equality} sourceId: ${sourceId}, resourceId: ${bfMediaNodeVideoGoogleDriveResourceId}`);
    }
}
```
