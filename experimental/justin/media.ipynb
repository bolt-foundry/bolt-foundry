```typescript
import { getJupyterCurrentViewer } from "infra/lib/jupyterUtils.ts"

const cv = await getJupyterCurrentViewer()
```

```typescript
import { BfMedia } from "packages/bfDb/models/BfMedia.ts";
import { BfMediaNodeVideo, BfMediaNodeVideoStatus, BfMediaNodeVideoRole } from "packages/bfDb/models/BfMediaNodeVideo.ts";
import { BfMediaNodeVideoGoogleDriveResource } from "packages/bfDb/models/BfMediaNodeVideoGoogleDriveResource.ts";
```

```typescript
const mediaId = "287d9da479eb44a2857036cf4036861e"
```

```typescript
const media = await BfMedia.find(cv, mediaId)
const targets = await media.queryTargetInstances(
  BfMediaNodeVideoGoogleDriveResource,
);
const resource = targets[0];

const video = await resource.queryTargetInstances(
    BfMediaNodeVideo,
    {
      status: BfMediaNodeVideoStatus.COMPLETED,
    }, {
      role: BfMediaNodeVideoRole.PREVIEW,
    }
);
await video[0].getFilePath();
```
